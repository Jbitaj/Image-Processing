```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_2 = cv.imread('images/Barbara.jpg', 0)
img_1 = cv.imread('images/AA.bmp', 0)
```

```python
img_2 = cv.resize(img_2, (img_1.shape[1], img_1.shape[0]))
```

تبدیل فوریه

```python
f1 =np.fft.fft2(img_1)
f2 =np.fft.fft2(img_2)
```

جابه جایی صفر در مرکز تبدیل فوریه

```python
f1_shifted = np.fft.fftshift(f1)
f2_shifted = np.fft.fftshift(f2)
```

قدر مطلق تبدیل فوریه برای رسمش 

```python
magnitude_spectrum1 = np.abs(f1_shifted)
magnitude_spectrum2 = np.abs(f2_shifted)
```

```python
plt.figure(figsize=(12, 8))

# تصویر اول
plt.subplot(2, 3, 1)
plt.imshow(img_1, cmap='gray')
plt.title('Original Image 1')
plt.axis('off')

# تصویر دوم
plt.subplot(2, 3, 2)
plt.imshow(img_2, cmap='gray')
plt.title('Resized Image 2')
plt.axis('off')

# تبدیل فوریه تصویر اول
plt.subplot(2, 3, 4)
plt.imshow(np.log(magnitude_spectrum1 + 1), cmap='gray')  # استفاده از لگاریتم برای نمایش بهتر
plt.title('Fourier Transform - Image 1')
plt.axis('off')

# تبدیل فوریه تصویر دوم
plt.subplot(2, 3, 5)
plt.imshow(np.log(magnitude_spectrum2 + 1), cmap='gray')  # استفاده از لگاریتم برای نمایش بهتر
plt.title('Fourier Transform - Image 2')
plt.axis('off')

plt.tight_layout()
plt.show()
```

طیف و فاز از تبدیل فوریه

```python
magnitude1 = np.abs(f1_shifted)
phase2 = np.angle(f2_shifted)
```

```python
magnitude2 = np.abs(f1_shifted)
phase1 = np.angle(f2_shifted)
```

```python
combined = magnitude1 * np.exp(1j * phase2)
```

```python
combined2 = magnitude2 * np.exp(1j * phase1)
```

```python
f_combined_shifted2 = np.fft.ifftshift(combined2)
image_combined2 = np.fft.ifft2(f_combined_shifted)
```

```python
image_combined_real = np.abs(image_combined)
```

```python
image_combined_real2 = np.abs(image_combined2)
```

```python
# فاز تصویر اول و طیف تصویر دوم را ترکیب می‌کنیم
magnitude2 = np.abs(f2_shifted)  # طیف از تصویر دوم
phase1 = np.angle(f1_shifted)    # فاز از تصویر اول

# ترکیب فاز تصویر اول با طیف تصویر دوم
combined_reversed = magnitude2 * np.exp(1j * phase1)

# انجام تبدیل فوریه معکوس
f_combined_reversed_shifted = np.fft.ifftshift(combined_reversed)
image_combined_reversed = np.fft.ifft2(f_combined_reversed_shifted)
image_combined_reversed_real = np.abs(image_combined_reversed)
```

```python
# نمایش تصاویر
plt.figure(figsize=(12, 8))

# تصویر اول
plt.subplot(2, 3, 1)
plt.imshow(img_1, cmap='gray')
plt.title('Original Image 1')
plt.axis('off')

# تصویر دوم
plt.subplot(2, 3, 2)
plt.imshow(img_2, cmap='gray')
plt.title('Resized Image 2')
plt.axis('off')

# طیف و فاز تصویر اول و دوم
plt.subplot(2, 3, 4)
plt.imshow(np.log(magnitude1 + 1), cmap='gray')
plt.title('Magnitude Spectrum of Image 1')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(np.angle(f2_shifted), cmap='gray')
plt.title('Phase of Image 2')
plt.axis('off')

# تصویر نهایی حاصل از ترکیب طیف و فاز
plt.subplot(2, 3, 6)
plt.imshow(image_combined_real, cmap='gray')
plt.title('Combined Image (Magnitude1 + Phase2)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
