```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
```

```python
color_image = cv.imread('images/woman-top-of-mountain-motivation.webp')
grayscale_image = cv.imread('images/mountaining.webp', 0)
```

```python
adjusted_color = exposure.rescale_intensity(color_image, in_range=(30, 180), out_range=(0, 255))
adjusted_grayscale = exposure.rescale_intensity(grayscale_image, in_range=(30, 180), out_range=(0, 255))
```

```python
rgb_image = cv.cvtColor(color_image, cv.COLOR_BGR2RGB)
adjusted_color = np.uint8(adjusted_color)
adjusted_rgb_image = cv.cvtColor(adjusted_color, cv.COLOR_BGR2RGB)
```

```python
plt.subplot(2,2,1)
plt.imshow(rgb_image)
plt.title('Original Color Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(grayscale_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(adjusted_rgb_image)
plt.title('Adjusted Color Image')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(adjusted_grayscale, cmap='gray')
plt.title('Adjusted Grayscale Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
